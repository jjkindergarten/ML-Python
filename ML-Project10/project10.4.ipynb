```python
import numpy as np
import math 
from functools import reduce
import random
import matplotlib.pyplot as plt
```

```python
def initial_value():
    x = [-1] + [random.choice([-1,1]) for i in range(10)]
    T = reduce(lambda m,n:m*n,x[1:11])
    wx = np.zeros((10,11))
    for i in range(10):
        wx[i] = np.random.uniform(1,-1,[1,11])
    wy = [np.random.uniform(1,-1) for i in range(11)]

    y = np.tanh(wx @ np.asarray(x))
    y = [-1] + list(y)

    S = np.tanh(np.asarray(wy) @ np.asarray(y))
    Es = 1/2 * (T - S)**2
    print (Es)
    return x, T, wx, wy, y, S, Es
```

```python
#learning function
def two_layer_network(x, T, wx, wy, y, S):
   #pde
    dwij = np.zeros((10,11))
    dwi = (S-T) * (1/np.cosh(np.asarray(wy) @ np.asarray(y)))**2 * np.asarray(y)
    for i in range(10):
        for j in range(11):
            dwij[i,j] = (S-T) * ((1/np.cosh(np.asarray(wy) @ np.asarray(y)))**2 * wy[i] *
                     (1/np.cosh(wx[i] @ np.asarray(x)))**2 * x[j])
    #learn
    wy = wy - 0.01 * dwi
    for i in range(10):
        wx[i] = wx[i] - 0.01 * dwij[i]
    #result
    y = np.tanh(wx @ np.asarray(x))
    y = [-1] + list(y)
    S = np.tanh(np.asarray(wy) @ np.asarray(y))
    
    E = 1/2 * (T - S)**2
    return wy, wx, y, S, E
```

```python
x1, T1, wx1, wy1, y1, S1, Es1 = initial_value()
```

```python
count = 0
E1 = Es1
Es1 = [Es1]
while E1 > 0.01:
    wy1, wx1, y1, S1, E1 = two_layer_network(x1, T1, wx1, wy1, y1, S1)
    Es1.append(E1)
    count = count + 1
    if count % 10000 == 0:
        print(count, E1)
    if count > 2000000:
        break
```
