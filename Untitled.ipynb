```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.optimize as optimization
```

```python
data = pd.read_excel('Book.xlsx')
mat1 = data.as_matrix()
data = np.transpose(data.as_matrix())
```

```python
x = data[1]
y = data[2]
plt.scatter(x,y)
plt.xlabel("Height (in)")
plt.ylabel("Weight (lbs)")
plt.show()
```

```python
selection = np.random.randint(0, len(data[0])-1, int(0.8*len(data[0])))
dataTrain = np.array([i for i in np.transpose(data) if i[0] in selection])
dataTest = np.array([i for i in np.transpose(data) if i[0] not in selection])
dataTrain = np.transpose(dataTrain)
dataTest = np.transpose(dataTest)
plt.scatter(dataTrain[1],dataTrain[2])
plt.title('Training Data')
plt.xlabel('Height (in)')
plt.ylabel("Weight (lbs)")
plt.show()
```

```python
plt.title('Test Data')
plt.scatter(dataTest[1],dataTest[2], c='red')
plt.xlabel('Height (in)')
plt.ylabel('Weight (lbs)')
plt.show()
```

```python
x = dataTrain[1]
y = dataTrain[2]
plt.scatter(x,y)
plt.xlabel('Height (in)')
plt.ylabel('Weight (lbs)')
def fun(x,a,b): return a + b*x
x0 = np.array([0.0,0.0])
optpar = optimization.curve_fit(fun, x,y,x0)
```

```python
a = optpar[0][0]
b = optpar[0][1]
plt.plot(x, a+b*x, "r")
plt.show()
```

```python
s = 0 
for n in range(np.size(x)):
    s = s + (y[n] - (a+b*x[n]))**2
```

```python
print("MSE=", s/np.size(x))
```
