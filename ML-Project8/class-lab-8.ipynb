### Dimensional Reduction ( SVD (Singular Value Decomposition) vs LLE (Local LInear Embedding))

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(1., 300., 150)

y1 = np.cos(-(x)/10) 
y2 = np.sin(-(x)/10)
y3 = x

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(y1, y2, y3)
plt.show()
```

###  Let us try SVD and rank 1 reduction for the spiral data introduced above

```python
from numpy.linalg import svd
M = np.column_stack((y1,y2,y3))
U,s,V = svd(M)
k = 1
Mnew = np.dot(U[:,:k],np.dot(np.diag(s[:k]),V[:k,:]))
A = np.hsplit(Mnew, 3)
y4 = A[0]
y5 = A[1]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(y1, y2,y3, 'ro')
ax.plot(y4,y5,0, 'bx')
plt.show()
```

### Projection of spiral data on to their firts principal component ( oscillating coordinates)

```python
plt.plot(Mnew[:,1],'bo')
plt.show()
```

## Coordinates should have a monotone increase along the spiral (that is done by LLE)

```python
from sklearn.manifold import LocallyLinearEmbedding
model1 = LocallyLinearEmbedding(n_neighbors= 3, n_components=1,  method='modified', eigen_solver='dense')
out1 = model1.fit_transform(M)

plt.plot(out1,'bx')
plt.show()
```
