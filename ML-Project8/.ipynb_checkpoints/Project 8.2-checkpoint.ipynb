```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from six.moves import urllib
from sklearn.datasets import fetch_mldata
from matplotlib import offsetbox
mnist = fetch_mldata('MNIST original')

from sklearn.model_selection import train_test_split

X = mnist["data"][::210]
y = mnist["target"][::210]
```

```python
from sklearn.manifold import MDS
mds = MDS(n_components=2, metric=True)
X_mds = mds.fit_transform(X).T
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X).T
```

```python
fig = plt.figure(figsize=(16, 8))
fig.add_subplot(121)
plt.scatter(X_mds[0], X_mds[1], c=y, cmap=plt.cm.rainbow)
fig.add_subplot(122)
plt.scatter(X_tsne[0], X_tsne[1], c=y, cmap=plt.cm.rainbow)
plt.show

    
```

I would say T-SNE do a better job. According to the color, points of diff number is more sparse in t-sne than mds.
