```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LassoLarsIC
```

```python
X =2 * np.random.rand(100, 1)
y = 5*np.sin(6*X) + 2*np.cos(X) +  np.random.randn(100, 1)
```

```python
plt.figure()
plt.scatter(X,y)
plt.show()
```

```python
def polyreg(x,y,deg): 
    
    poly_features = PolynomialFeatures(degree = deg, include_bias =False)
    X_poly = poly_features.fit_transform(X)
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly,y)
    y_predict = lin_reg.predict(X_poly)
    model_bic = LassoLarsIC(criterion='bic')
    mse = mean_squared_error(y, y_predict)
    
    X_new=np.linspace(0, 2, 100).reshape(100, 1)
    X_new_poly = poly_features.fit_transform(X_new)
    y_new = lin_reg.predict(X_new_poly)
    
    return mse, X_new, y_new
```

```python
plt.figure(num='linear_reg',figsize=(18,6))
           
for subplot in range(241,249):
    mse, X_new, y_new = polyreg(X,y,(subplot-239))
    
    plt.subplot(subplot)
    plt.plot(X, y, "b.")
    plt.plot(X_new, y_new, "r-", linewidth=2)
    plt.xlabel("$x_1$", fontsize=18)
    plt.ylabel("$y$", rotation=0, fontsize=18)
    plt.legend(loc="upper left", fontsize=14)
    plt.axis([-0.1, 2.1, -8, 8])
    plt.title("degree = " + str(subplot-239) + " mse = " + str(round(mse,2)))
```
