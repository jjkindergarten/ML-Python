```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
ones = []
zeros = []
test = []
```

```python
for i in range(40):
    one = Image.open("training_ones/"+str(i)+".png").convert('L')
    zero = Image.open("training_zeros/"+str(i)+".png").convert('L')
    ones.append(np.asarray(one))
    zeros.append(np.asarray(zero))
```

```python
for i in range(10):
    t = Image.open("test/"+str(i)+".png").convert('L')
    test.append(np.asarray(t))
```

```python
f,ax = plt.subplots(1,2)
```

```python
zeros_avg = np.average(zeros,axis=0)
ones_avg = np.average(ones,axis = 0 )
```

```python
ax[0].imshow(zeros_avg,cmap = 'gray')
ax[0].set_title("Average Zero")
ax[1].imshow(ones_avg,cmap='gray')
ax[1].set_title("Average One")
```

```python
plt.show()
```

```python
counter = 0 
```

```python
if np.linalg.norm(test[counter]-zeros_avg)<np.linalg.norm(test[counter]-ones_avg):
    print('i think this is a zero')
    plt.imshow(zeros_avg-test[counter],cmap='gray')
else:
    print('i think this is a one')
    plt.imshow(ones_avg-test[counter],camp='gray')
```

```python
plt.title(counter)
plt.show()
```

```python
counter += 1
```

```python
if counter >= len(test):
    counter = 0
```

```python
ones_distances = [np.linalg.norm(one-test[counter])for one in ones]
zeros_distances = [np.linalg.norm(zero-test[counter])for zero in zeros]
```

```python
if min(zeros_distances)< min(ones_distances):
    print ('i think this is a zero')
    plt.imshow(zeros[np.argmin(zeros_distances)] - test[counter], cmap = 'gray')
    
else:
    print('i think this is a one')
    plt.imshow(ones[np.argmin(ones_distances)]-test[counter],cmap='gray')
```

```python
plt.title(counter)
plt.show()
```

```python
# hw1 question 1
## pca
ones_pca = []
zeros_pca = []
from sklearn.decomposition import PCA 
pca = PCA(n_components = 2)
```

```python
for i in range(40):
    one_pca = pca.fit_transform(ones[i])
    zero_pca = pca.fit_transform(zeros[i])
    ones_pca.append(np.asarray(one_pca))
    zeros_pca.append(np.asarray(zero_pca))
```

```python
test_pca = []
for i in range(10):
    t_pca = pca.fit_transform(test[i])
    test_pca.append(np.asarray(t_pca))
```

```python
ones_pca_distance = [np.linalg.norm(one_pca - test_pca[1]) for one_pca in ones_pca]
zeros_pca_distance = [np.linalg.norm(zero_pca - test_pca[i]) for zero_pca in zeros_pca]
```

```python
#k = 1
if min(zeros_pca_distance)< min(ones_pca_distance):
    print ('i think this is a zero')
    plt.imshow(zeros_pca[np.argmin(zeros_pca_distance)] - test_pca[1], cmap = 'gray')
    
else:
    print('i think this is a one')
    plt.imshow(ones_pca[np.argmin(ones_pca_distance)]-test_pca[counter],cmap='gray')
```

```python
plt.title(counter)
plt.show()
```
