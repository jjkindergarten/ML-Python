Decision tree training code

```python
# function for the grid to use for the decision tree results visualization
def get_grid(data):
    x_min, x_max = data[:, 0].min() - 1, data[:, 0].max() + 1
    y_min, y_max = data[:, 1].min() - 1, data[:, 1].max() + 1
    return np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
```

```python
def decisiontree_visualization(data,labels):
    from sklearn.tree import DecisionTreeClassifier
    # parameter min_samples_leaf is the minimal number of the elements
    # for the node splitting
    clf_tree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=17)

    # tree training 
    clf_tree.fit(data, labels)

    # code to show the data splitting 
    xx, yy = get_grid(data)
    predicted = clf_tree.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.pcolormesh(xx, yy, predicted, cmap='autumn')
    plt.scatter(data[:, 0], data[:, 1], c=labels, s=100, 
    cmap='autumn', edgecolors='black', linewidth=1.5);
```

Code to plot the tree using png file (you can just open png file itself for a better quality) 

```python
# let's rotate the samples in the same angles so that we rotate the coordinate system
#use the rotation matrix
def rotate(data, angle):
    rotation_matrix = np.array([np.cos(angle*math.pi/180.0), -np.sin(angle*math.pi/180.0), np.sin(angle*math.pi/180.0),
                       np.cos(angle*math.pi/180.0)]).reshape(2,2)
    rotated_data = [np.dot(rotation_matrix, np.transpose(sample)) for sample in data]
    rotated_data = np.asarray(rotated_data)
    return rotated_data    
```

```python
rotated_train_data = rotate(train_data, 30)
```

```python
plt.figure(num='rotation & decision tree',figsize=(20,45))
           
for sub in range(1,37):
    
    plt.subplot(9,4, sub)
    rotated_train_data = rotate(train_data, (sub-1)*10)
    decisiontree_visualization(rotated_train_data,train_labels)
```

Generate data about the function $f(x) = e^{-x^2} + 1.5 e^{-(x-2)^2}$ to train a decison tree for regression 

```python
n_train = 150        
n_test = 1000       
noise = 0.1

def f(x):
 x = x.ravel()
 return np.exp(-x ** 2) + 1.5 * np.exp(-(x - 2) ** 2)

def generate(n_samples, noise):
 X = np.random.rand(n_samples) * 10 - 5
 X = np.sort(X).ravel()
 y = np.exp(-X ** 2) + 1.5 * np.exp(-(X - 2) ** 2) + np.random.normal(0.0, noise, n_samples)
 X = X.reshape((n_samples, 1))
 return X, y

X_train, y_train = generate(n_samples=n_train, noise=noise)
X_test, y_test = generate(n_samples=n_test, noise=noise)

from sklearn.tree import DecisionTreeRegressor

reg_tree = DecisionTreeRegressor(max_depth=5, random_state=17)

reg_tree.fit(X_train, y_train)
reg_tree_pred = reg_tree.predict(X_test)

plt.figure(figsize=(10, 6))
plt.plot(X_test, f(X_test), "b")
plt.scatter(X_train, y_train, c="b", s=20)
plt.plot(X_test, reg_tree_pred, "g", lw=2)
plt.xlim([-5, 5])
plt.title("Decision tree regressor, MSE = %.2f" % np.sum((y_test - reg_tree_pred) ** 2))
plt.show()
```

In the last cell of the code compare performance of the decision tree with a polynomial regression.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def polyreg(X_train, y_train, X_test, y_test, deg): 
    
    poly_features = PolynomialFeatures(degree = deg, include_bias =False)
    X_poly = poly_features.fit_transform(X_train)
    lin_reg = LinearRegression()
    lin_reg.fit(X_poly,y_train)

    
    X_test_poly = poly_features.fit_transform(X_test)
    y_predict = lin_reg.predict(X_test_poly)
    mse = np.sum((y_test - y_predict) ** 2)
    
    return mse, y_predict
```

```python
#let me first find the suitable parameters for the ploy regression

plt.figure(num='linear_reg',figsize=(18,6))
           
for subplot in range(241,249):
    mse, y_predict = polyreg(X_train, y_train, X_test, y_test, (subplot-237))
    
    plt.subplot(subplot)
    plt.plot(X_train, y_train, "b.")
    plt.plot(X_test, y_predict, "r-", linewidth=2)
    plt.xlabel("$x_1$", fontsize=18)
    plt.ylabel("$y$", rotation=0, fontsize=18)
    plt.legend(loc="upper left", fontsize=14)
    plt.axis([min(X_train), max(X_train), min(y_train), max(y_train)])
    plt.title("degree = " + str(subplot-237) + " mse = " + str(round(mse,2)))
```

<font size = 3> the curve of ploynominal regression looks more smooth. Moreover, if we count each node of a tree as a parameter, then the decision tree has too much parameters. On the other hand, polynomianl regression can achieve a similar mse with degree of '10'. Hence, i think ploynominal is better for regression.
